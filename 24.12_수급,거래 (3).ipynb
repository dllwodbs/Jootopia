```python
import pandas as pd
from pykrx import stock
import pandas as pd
from tqdm import tqdm
import time
```

```python
df_k200 = pd.read_csv('24.12_kospi200 -회사명,종목코드.csv', dtype={'종목코드' : str})
tickers = df_k200["종목코드"][:10].tolist()
```

```python
df_k200
```

```python
for i in range(len(df_k200)):
    data = stock.get_market_ohlcv_by_date("20150101", "20241231", df_k200['종목코드'][i])
    data['종목코드'] = df_k200['종목코드'][i]
    data.to_csv(f'주식data/{df_k200['종목명'][i]}.csv')
    print(i)
    time.sleep(2)
```

```python
tickers
```

```python
start = "20150101"
end   = "20241231"


all_vol_data = []

for ticker in tqdm(tickers):
    try:
        # ② 투자자별 '순매수 거래대금(원)' 데이터 불러오기
        df_vol = stock.get_market_trading_volume_by_date(
            fromdate=start,
            todate=end,
            ticker=ticker
        )

        # ③ 필요한 컬럼만 추출
        vol = df_vol[["기관합계", "개인", "외국인합계"]].copy()
        vol.columns = ["기관_순매수_수량", "개인_순매수_수량", "외국인_순매수_수량"]

        # ④ ticker 정보 추가
        vol["ticker"] = ticker

        all_vol_data.append(vol)
        time.sleep(3)

    except Exception as e:
        print(f"Error in {ticker}: {e}")

# ⑤ 모든 종목 데이터 합치기
result2 = pd.concat(all_vol_data)
result2 = result2.sort_index()
result2.index.name = "date"

print(result2.head())
print(result2.tail())
```
