```python
import os
```

```python
os.listdir('../주식data/주식data')
```

```python
df = pd.DataFrame()

for list in os.listdir('../주식data/주식data'):
    df2 = pd.read_csv(f'../주식data/주식data/{list}', dtype={'종목코드':str})
    df2.rename(columns={'종목코드':'ticker', '날짜':'date'}, inplace=True)
    df = pd.concat([df, df2])
df = df.reset_index(drop=True)
df
```

```python
price_df = df.sort_values(['ticker', 'date'])

price_df['ret_1d']  = price_df.groupby('ticker')['종가'].pct_change(1)
price_df['ret_5d']  = price_df.groupby('ticker')['종가'].pct_change(5)
price_df['ret_20d'] = price_df.groupby('ticker')['종가'].pct_change(20)
```

```python
price_df.reset_index(drop=True, inplace=True)
```

```python
price_df.to_csv('단기모멘텀.csv', index=False)
```
