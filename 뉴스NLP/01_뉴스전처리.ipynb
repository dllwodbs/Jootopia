```python
import pandas as pd
import os
```

```python
kospi200 = pd.read_csv('../24.12_kospi200 -회사명,종목코드.csv', dtype={'종목코드' : str})
kospi200
```

```python
company_list = os.listdir('../주식뉴스')
company_list
```

```python
for company in os.listdir('../주식뉴스'):
    df = pd.DataFrame()
    file_list = os.listdir(f'../주식뉴스/{company}')
    for file in file_list:
        df2 = pd.read_excel(f'../주식뉴스/{company}/{file}')
        df = pd.concat([df, df2])
    df = df[['일자', '제목']]
    df['ticker'] = kospi200[kospi200['종목명'] == company]['종목코드'].values[0]
    df.sort_values('일자', inplace=True)
    df.reset_index(drop=True, inplace=True)
    df.to_csv(f'../newsdata/{company}.csv', index=False)
```

```python
df = pd.DataFrame()

for file in file_list:
    df2 = pd.read_excel(f'../주식뉴스/LG/{file}')
    df = pd.concat([df, df2], ignore_index=True)
    df['ticker'] = kospi200[kospi200['종목명'] == 'LG']['종목코드'].values[0]
```

```python
df.info()
```

```python
df2 = df.copy()
```

```python
df2 = df2[['일자', '제목', '키워드', '특성추출(가중치순 상위 50개)', '통합 분류1']]
```

```python
# LG 관련 회사명 제외 리스트
exclude = company_list[25:32] + ['LG생건', 'LG U+']

def is_lg_only_simple(title):
    return ("LG" in title) and not any(x in title for x in exclude)

df_lg_only = df2[df2["제목"].apply(is_lg_only_simple)]
```

```python
df_lg_only.reset_index(drop=True, inplace=True)
```

```python
# ESG 키워드 포함된 뉴스 인덱스 추출
esg_list = list()

for idx, data in enumerate (df_lg_only['키워드']):
    # print(idx)
    # print(data)
    if 'ESG' in data:
        esg_list.append(idx)
```
